```python
from datetime import datetime
import boto3
import json
import random
import time

### conectando ao kinesis ###
cliente = boto3.client('kinesis',
aws_access_key_id="###################",
aws_secret_access_key="#######################",
region_name = "sa-east-1"
)

### criando produtor de Battery Temperature ###
t= True
id=0
while t:
    dados = round(random.uniform(20,25),4)  
    id +=1
    if id == 1500:
        t = False
    registro = {'id':str(id),'data':str(dados), 'type':"temperature", 'timestamp' : str(datetime.now())} 
   
    cliente.put_record(       #gerando resposta do kinesis
    StreamName = 'Engdados_projeto2',
    Data = json.dumps(registro),     #convertendo meus ados para o formato json
    PartitionKey='02'                # chave de partição que gerará um hash. Se outro produtor tiver a mesma chavem gerará o memso hash e irá pro mesmo shard no kinesis
    )
    time.sleep(0.3)
    print (registro)
```
